```python
df=pd.read_excel("./Sample-data-ML-13AUG2020.xlsx")
print(df.columns)
df=df[['TICKET_SIZE','AUTO_MANDATE_FLAG','FUP_DATE','ANNULIZED',
       'FUP_BUCKET', 'BOOKING_FREQUENCY', 'CHANNEL',
       'PERSISTENCY_FLAG', 'TERM_MONTH', 'FUP_MONTH',
       'PAYMETHOD','CLEAN_STATE','P_PRODUCT_TYPE', 'PRODUCT_NAME', 'AGENT_STATUS']]
df['TICKET_SIZE']=df.TICKET_SIZE.astype(float)
df['TICKET_SIZE']=df.TICKET_SIZE.astype(float)
df['ANNULIZED']=df.ANNULIZED.astype(float)
df['BOOKING_FREQUENCY']=df.BOOKING_FREQUENCY.astype(float)
```

```python
x=df.FUP_DATE.dt.month
y=df.CLEAN_STATE
# plt.figure(figsize=(30,10))
e=df.groupby([x,y]).count()
e.iplot(kind='bar')
```

```python
# cf.getThemes()
df[(df.CLEAN_STATE=='Andhra Pradesh') & (df.FUP_DATE.dt.month==2)]
```

```python
y=df.ANNULIZED
df.iplot(kind='bar',x='CLEAN_STATE',y='ANNULIZED')
```

```python
x=df.FUP_DATE.dt.year
y=df.CLEAN_STATE
# plt.figure(figsize=(30,10))
e=df.groupby([y,x]).count()
e.iplot(kind='bar')
```

```python
df[(df.CLEAN_STATE=='Andhra Pradesh') & (df.FUP_DATE.dt.year==2018)]
```

```python
x=df.AGENT_STATUS
y=df.CLEAN_STATE
# plt.figure(figsize=(30,10))
e=df.groupby([y,x]).count()
e.iplot(kind='bar')
```

```python
x=df.AGENT_STATUS
y=df.CLEAN_STATE
z=df.FUP_DATE.dt.month
s=df.FUP_DATE.dt.year
# plt.figure(figsize=(30,10))
e=df.groupby([y,x,z,s]).count()
e.iplot(kind='bar')
```

```python
x=df.CHANNEL
y=df.CLEAN_STATE
df.groupby([x,y]).count().iplot(kind='bar')
```
